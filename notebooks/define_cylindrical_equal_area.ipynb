# Define Cylindrical Equal Area Projection for cartopy

```python
import cartopy.crs as ccrs
import math

class CylindricalEqualArea(ccrs._RectangularProjection):
    def __init__(self, central_longitude=0.0, true_scale_latitude=0.0, globe=None):
        proj4_params = [
            ('proj', 'cea'), 
            ('lon_0', central_longitude), 
            ('lat_ts', true_scale_latitude)
        ]
        if globe == None:
            globe = ccrs.Globe() #semimajor_axis=math.degrees(1))
        super(CylindricalEqualArea, self).__init__(proj4_params, 180,
                                                 89., globe=globe)

    @property
    def threshold(self):
        return 0.5
```

```python
CylindricalEqualArea(true_scale_latitude=30.)
#math.degrees(1)
```
